**Question 5:** Write a Python program to calculate the factorial of a number.

```python
# Your code here

def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

number = int(input("Enter number : "))
print(f"Factorial of {number} is -> {factorial(number)}")
```

**Question 16:** Write a Python program to find the sum of the first n natural numbers.

```python
# Your code here
def sum_of_n(num):
    if num == 0:
        return 0
    
    return num + sum_of_n(num - 1)

number = int(input("Enter a number : "))
print(f"Sum of first {number} natural numbers is {sum_of_n(number)}")
```

**Question 17:** Write a Python program to check if a year is a leap year.

```python
def leap(year):
    if year % 4 != 0:
        return f"{year} is not a leap year"
    elif year % 100 != 0:
        return f"{year} is a leap year"
    elif year % 400 == 0:
        return f"{year} is a leap year"
    else:
        return f"{year} is not a leap year"

year = int(input("Enter the year: "))
print(leap(year))
```

**Question 18:** Write a Python program to reverse a string.

```python
# Your code here
def reverse(string):
    left = 0
    right = len(string) - 1

    while(left < right):
        string[left], string[right] = string[right], string[left]
        left += 1
        right -= 1
    return string

string = "HELLO"
print(f"Original string : {string} and reversed string is {"".join(reverse(list(string)))}")
```

**Question 19:** Write a Python program to check if a string is a palindrome.

```python
# Your code here
def is_palindrome(string):
    left = 0
    right = len(string) - 1
    
    while left < right:
        if string[left] != string[right]:
            return False
        left += 1
        right -= 1
    return True

string = "AABBAA"
string2 = "ABCBAA"

print(f"{string} is palindrome -> {is_palindrome(string)}")
print(f"{string2} is palindrome -> {is_palindrome(string2)}")
```
